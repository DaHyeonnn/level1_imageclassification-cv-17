```python
import os
os.getcwd()
```

```python
import csv
import pandas as pd
test_dir = '../input/data/eval/'
submission = pd.read_csv(os.path.join(test_dir, 'info.csv'))

# 비교할 두 개의 csv 파일 경로 설정
mask_file = './sm_result/mask.csv' #상민넷
gender_file = './sm_result/gender.csv'  # 상민넷
age_file = './Age_new/sh_eff3_cross_adam_real2/output_e20.csv'

ans=0
all_predictions=[]
with open(mask_file, 'r') as file1, open(gender_file, 'r') as file2, open(age_file, 'r') as file3:
    reader1 = csv.reader(file1)
    reader2 = csv.reader(file2)
    reader3 = csv.reader(file3)
       
    for row1, row2,row3 in zip(reader1, reader2,reader3):
        if row1[1]!='ans' and row2[1]!='ans' and row3[1]!='ans':
            ans=6*int(row1[1])+3*int(row2[1])+int(row3[1])
            all_predictions.append(ans)
            
submission['ans'] = all_predictions
save_dir = './Age_new/'
submission.to_csv(os.path.join(save_dir, 'sh_eff3_cross_adam_real2.csv'), index=False)
print('test inference is done!')
```
